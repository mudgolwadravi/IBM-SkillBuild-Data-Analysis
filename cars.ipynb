```python
def remove_outlier(col_name):
    sorted(col_name)
    Q1,Q3=col_name.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_bound=Q1-1.5*IQR
    upper_bound=Q3+1.5*IQR
    return lower_bound,upper_bound
```

```python
import numpy as np
low,high=remove_outlier(df1['mpg'])
```

```python
df1['mpg']=np.where(df1['mpg']>high,high,df1['mpg'])
```

```python
df1['mpg']=np.where(df1['mpg']<low,low,df1['mpg'])
```

```python
sns.boxplot(y='mpg',data=df1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
sns.histplot(data=df1, x='mpg', bins=15)
plt.title('Distribution of Miles Per Gallon (MPG)')
plt.xlabel('MPG')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df1, x='wt', y='mpg')
plt.title('Weight vs MPG')
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('MPG')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df1, x='cyl', y='mpg')
plt.title('MPG Distribution by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df1, x='cyl', y='mpg', palette='muted')
plt.title('Distribution of MPG by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Miles Per Gallon')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df1, x='am', y='hp', palette='coolwarm')
plt.title('Horsepower Distribution by Transmission Type')
plt.xlabel('Transmission (0 = Automatic, 1 = Manual)')
plt.ylabel('Horsepower')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df1, x='cyl', palette='viridis')
plt.title('Count of Cars by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```
